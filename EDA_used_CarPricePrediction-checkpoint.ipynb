# Download and Import the csv Data as Pandas DataFrame

```python
import os

download_dir = "./Data/"

download_url = "https://raw.githubusercontent.com/aravind9722/datasets-for-ML-projects/main/cardekho_dataset.csv"

os.makedirs(download_dir,exist_ok=True)

filename = os.path.basename(download_url) # fetching the path from where we downloaded - cardekho_dataset.csv

download_file_path = os.path.join(download_dir,filename) # join the path

print(filename)

print(download_file_path)

urllib.request.urlretrieve(download_url, download_file_path) # retrieving entire file(pulling the particular data from web url)

df = pd.read_csv(download_file_path, index_col=[0])
```

```python
print(df['max_power'].dtype != 'O')
print(df['transmission_type'].dtype == 'O') # checking whether datatype is object or not
```

```python
# define numerical and categorical columns

# if datatype is not object then it will devide features into numeric_features otherwise categorical features

numeric_features = [feature for feature in df.columns if df[feature].dtype != 'O'] # list comprehension
categorical_features = [feature for feature in df.columns if df[feature].dtype == 'O']

# print columns

print('We have {} numerical features : {}'.format(len(numeric_features), numeric_features))
print('\nWe have {} categorical features : {}'.format(len(categorical_features), categorical_features))
```

# Feature Information

###### car_name: Car's full name, which includes brand and specific model name.
###### brand: Brand Name of the particular car. 
###### model: Exact model name of the car of a particular brand. 
###### seller_type: Which Type of seller is selling the used car. 
###### fuel_type: Fuel used in the used car, which was put up on sale. 
###### transmission_type : transmission used in the used car, which was put on sale. 
###### vehicle_age : The count of years since car was bought. 
###### mileage: It is the number of kilometer the car runs per litre. 
###### engine: It is the engine capacity in cc(cubic centimeters). 
###### max_power: Max power it produces in BHP. 
###### seats: Total number of seats in car. 
###### selling_price: The sale price which was put up on website.


```python
# checkong how much percentage of the values are there wrt to each categoris in every categorical features
# proportion of count data on categorical columns

for col in categorical_features:
    print(df[col].value_counts(normalize=True) * 100) # * 100 means that many number of %
    print('---------------------------')
```

# Univariate Analysis

The term univariate analysis refers to the analysis of one variable prefix "uni" means "one". The purpose of univariate analysis is to understand the distribution of values for a single variable.

### Numerical Features

```python
plt.figure(figsize=(15,15))
plt.suptitle('Univariate Analysis of Numerical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.0)

for i in range(0, len(numeric_features)):
    plt.subplot(5, 3, i+1) # max 5 rows and 3 columns and index given where the plot needs to be displayed
    sns.kdeplot(x=df[numeric_features[i]], shade=True, color='b') # kernel density estimator plot - able to create a pdf function - distribution of any numerical features
    plt.xlabel(numeric_features[i])
    plt.tight_layout() #precisly display all plots
```

```python
# To check outliers present

sns.boxplot(df['max_power'])
```

```python
sns.boxplot(df['km_driven'])
```

### Report

km_driven,max_power,selling_price and engine are right skewed and positively skewed.

Outliers in km_driven, engine, selling_price and max_power.

## Categorical Features

```python
# Categorical columns

plt.figure(figsize=(20,15))
plt.suptitle('Univariate Analysis of Categorical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.0)
cat1 = ['brand', 'seller_type','fuel_type', 'transmission_type']

for i in range(0, len(cat1)):
    plt.subplot(2,2, i+1)
    sns.countplot(x=df[cat1[i]])
    plt.xlabel(cat1[i])
    plt.xticks(rotation=45)
    plt.tight_layout()
```

# Multivariate Analysis

Multivariate Analysis is the analysis of more than one variable

### Check Multicollinearity in Numerical Features

Multicollinearity - within independent features, two features are highly correlated to each other

```python
df[(list(df.columns)[1:])].corr() # values ranges between -1 to 1

# km_driven is 0.33 % positively correlated with vehicle age which means
# when 1 unit increase in vehicle age may increase 0.33 unit of km_driven

# when cor is negative then when vehicle age increase and mileage will get reduced.
```

```python
plt.figure(figsize = (15,10))
sns.heatmap(df.corr(), cmap='CMRmap', annot=True)
plt.show()
```
